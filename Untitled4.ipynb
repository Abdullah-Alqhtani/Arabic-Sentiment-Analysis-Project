```python
%%writefile preprocessing.py
import re
from typing import List, Union


class ArabicPreprocessor:
    """
    A preprocessing pipeline for Arabic Twitter text.

    Each method is a standalone transformation step.
    The `preprocess` method chains them together.
    """

    _DIACRITICS_PATTERN = re.compile(r"[\u064B-\u065F\u0670]")
    _URL_PATTERN = re.compile(r"https?://\S+|www\.\S+")
    _MENTION_PATTERN = re.compile(r"@\w+")
    _NON_ARABIC_PATTERN = re.compile(r"[^\u0600-\u06FF\u0750-\u077F\u08A0-\u08FF\s]")
    _EMOJI_PATTERN = re.compile(
        "["
        "\U0001F300-\U0001F5FF"
        "\U0001F600-\U0001F64F"
        "\U0001F680-\U0001F6FF"
        "\U0001F700-\U0001F77F"
        "\U0001F780-\U0001F7FF"
        "\U0001F800-\U0001F8FF"
        "\U0001F900-\U0001F9FF"
        "\U0001FA00-\U0001FAFF"
        "]+",
        flags=re.UNICODE,
    )
    _REPEATED_PATTERN = re.compile(r"(.)\1{2,}")
    _WHITESPACE_PATTERN = re.compile(r"\s+")

    def remove_diacritics(self, text: str) -> str:
        return self._DIACRITICS_PATTERN.sub("", text)

    def normalize_alef(self, text: str) -> str:
        return re.sub(r"[أإآٱ]", "ا", text)

    def normalize_teh_marbuta(self, text: str) -> str:
        return text.replace("ة", "ه")

    def remove_urls(self, text: str) -> str:
        return self._URL_PATTERN.sub(" ", text)

    def remove_mentions(self, text: str) -> str:
        return self._MENTION_PATTERN.sub(" ", text)

    def remove_hashtags(self, text: str) -> str:
        return text.replace("#", "")

    def remove_punctuation_and_emojis(self, text: str) -> str:
        text = self._EMOJI_PATTERN.sub(" ", text)
        text = self._NON_ARABIC_PATTERN.sub(" ", text)
        return text

    def remove_repeated_characters(self, text: str) -> str:
        return self._REPEATED_PATTERN.sub(r"\1", text)

    def normalize_whitespace(self, text: str) -> str:
        return self._WHITESPACE_PATTERN.sub(" ", text).strip()

    def tokenize(self, text: str) -> List[str]:
        return [token for token in text.split() if token]

    def preprocess(self, text: str, tokenize: bool = True) -> Union[List[str], str]:
        if not isinstance(text, str):
            text = "" if text is None else str(text)

        cleaned = text
        cleaned = self.remove_urls(cleaned)
        cleaned = self.remove_mentions(cleaned)
        cleaned = self.remove_hashtags(cleaned)
        cleaned = self.remove_diacritics(cleaned)
        cleaned = self.normalize_alef(cleaned)
        cleaned = self.normalize_teh_marbuta(cleaned)
        cleaned = self.remove_punctuation_and_emojis(cleaned)
        cleaned = self.remove_repeated_characters(cleaned)
        cleaned = self.normalize_whitespace(cleaned)

        if tokenize:
            return self.tokenize(cleaned)
        return cleaned
```

```python
from collections import Counter
from typing import List
import math


class NaiveBayesClassifier:
    def __init__(self):
        self.class_counts = Counter()
        self.word_counts = {}
        self.vocab = set()
        self.total_words = {}

    def train(self, texts: List[List[str]], labels: List[str]):
        for tokens, label in zip(texts, labels):
            self.class_counts[label] += 1

            if label not in self.word_counts:
                self.word_counts[label] = Counter()
                self.total_words[label] = 0

            for word in tokens:
                self.word_counts[label][word] += 1
                self.total_words[label] += 1
                self.vocab.add(word)

    def _log_prior(self, label: str) -> float:
        total = sum(self.class_counts.values())
        return math.log(self.class_counts[label] / total)

    def _log_likelihood(self, word: str, label: str) -> float:
        word_count = self.word_counts[label][word] + 1
        total = self.total_words[label] + len(self.vocab)
        return math.log(word_count / total)

    def predict(self, tokens: List[str]) -> str:
        scores = {}

        for label in self.class_counts:
            score = self._log_prior(label)

            for word in tokens:
                score += self._log_likelihood(word, label)

            scores[label] = score

        return max(scores, key=scores.get)
```

```python
%%writefile evaluation.py
from typing import List, Tuple

def accuracy(predictions: List[str], gold: List[str]) -> float:
    correct = sum(p == g for p, g in zip(predictions, gold))
    return correct / len(gold) if gold else 0.0

def precision_recall_f1(
    predictions: List[str],
    gold: List[str],
    positive_label: str = 1
) -> Tuple[float, float, float]:
    tp = sum((p == positive_label and g == positive_label) for p, g in zip(predictions, gold))
    fp = sum((p == positive_label and g != positive_label) for p, g in zip(predictions, gold))
    fn = sum((p != positive_label and g == positive_label) for p, g in zip(predictions, gold))

    precision = tp / (tp + fp) if (tp + fp) else 0.0
    recall = tp / (tp + fn) if (tp + fn) else 0.0
    f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) else 0.0

    return precision, recall, f1
```

```python
from evaluation import accuracy, precision_recall_f1

acc = accuracy(predictions_clean, test_labels)
precision, recall, f1 = precision_recall_f1(predictions_clean, test_labels, positive_label=1)

print("Accuracy:", acc)
print("Precision:", precision)
print("Recall:", recall)
print("F1:", f1)
```
